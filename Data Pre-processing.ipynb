```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\SHON SEBASTIAN\OneDrive\Desktop\New folder\Social_Media_Advertising.csv")
df.head()
```

```python
sns.countplot(x='Campaign_Goal', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='Channel_Used', y='ROI', data=df)
plt.show()

sns.scatterplot(x='Impressions', y='Clicks', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
df.groupby('Channel_Used')['ROI'].mean()
```

```python
df.groupby('Campaign_Goal')['Clicks'].sum()
```

```python
df.groupby('Customer_Segment')['Conversion_Rate'].mean()
```

```python
df.groupby('Target_Audience')['Engagement_Score'].mean()
```

```python
df.groupby('Target_Audience')['Engagement_Score'].mean()
```

```python
pivot_roi = pd.pivot_table(
    df,
    values='ROI',
    index='Channel_Used',
    aggfunc='mean'
)

print(pivot_roi)
```

```python
pivot_conversion = pd.pivot_table(
    df,
    values='Conversion_Rate',
    index='Channel_Used',
    columns='Customer_Segment',
    aggfunc='mean'
)

print(pivot_conversion)
```

```python
pivot_clicks = pd.pivot_table(
    df,
    values='Clicks',
    index='Campaign_Goal',
    aggfunc='sum'
)

print(pivot_clicks)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y', errors='coerce')
```

```python
df['Campaign_Month'] = df['Date'].dt.month
```

```python
pivot_month = pd.pivot_table(
    df,
    values='Impressions',
    index='Campaign_Month',
    aggfunc='sum'
)

print(pivot_month)
```

```python
corr_matrix = df.corr(numeric_only=True)
print(corr_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Between Numerical Variables")
plt.show()
```
