```python
print("Missing values in each column:\n")
print(df.isnull().sum())

print("\nTotal Missing Values :", df.isnull().sum().sum())
```

```python
df = df.dropna()
```

```python
print("\nAfter removing missing values:\n")
print(df.isnull().sum())
```

```python
print("Duplicate Rows :", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
print("\nAfter removing duplicates :")
print("Duplicate Rows :", df.duplicated().sum())
```

```python
(df.dtypes)
```

```python
df['Campaign_ID'] = df['Campaign_ID'].astype('int64')
df['Clicks'] = df['Clicks'].astype('int64')
df['Impressions'] = df['Impressions'].astype('int64')

print(df.dtypes)
```

```python
print(df.head())
```

```python
df['CTR'] = df['Clicks'] / df['Impressions']
```

```python
df['CPC'] = df['Acquisition_Cost'] / df['Clicks']
```

```python
df['Conversion_Per_Click'] = df['Conversion_Rate'] / df['Clicks']
```

```python
df['Campaign_Month'] = df['Date'].dt.month
```

```python
df['Campaign_Day'] = df['Date'].dt.day_name()
```

```python
print(df.head())
```

```python
high_roi = df[df['ROI'] > 5]
print(high_roi.head())
```

```python
instagram_data = df[df['Channel_Used'] == 'Instagram']
print(instagram_data.head())
```

```python
young_campaigns = df[df['Target_Audience'] == 'Women 18-24']
print(young_campaigns.head())
```
